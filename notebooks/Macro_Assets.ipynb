```python
%pip install yfinance
import yfinance as yf

# Ticker lists for 10 major countries (Equity ETFs)
us_tickers = ["SPY"]
china_tickers = ["MCHI"]
japan_tickers = ["EWJ"]
uk_tickers = ["EWU"]
germany_tickers = ["EWG"]
canada_tickers = ["EWC"]
india_tickers = ["INDA"]
france_tickers = ["EWQ"]
australia_tickers = ["EWA"]
brazil_tickers = ["EWZ"]

# Bond ETFs (popular selections)
bond_tickers = ["AGG", "BND", "TLT", "BNDX", "VCIT", "MUB", "VCSH"]

# Commodity ETFs (popular selections)
commodity_tickers = ["GLD", "SLV", "PDBC", "FTGC", 'BNO', 'USO', 'UNG']

# Combine all ticker lists into one for download
all_stock_tickers = (
    us_tickers +
    china_tickers +
    japan_tickers +
    uk_tickers +
    germany_tickers +
    canada_tickers +
    india_tickers +
    france_tickers +
    australia_tickers +
    brazil_tickers
)
all_bond_tickers = bond_tickers
all_commodity_tickers = commodity_tickers
all_tickers = all_stock_tickers + all_bond_tickers + all_commodity_tickers

# Function to download full historical data for each ticker
def download_full_history(tickers):
    # Download full history data for all tickers
    data = yf.download(tickers, period="max")  # Downloads full historical data
    return data

# Download full historical data for all tickers
full_history_data = download_full_history(all_tickers)

# Display data
#close_prices = full_history_data['Adj Close'] # no adj close price?
close_prices1 = full_history_data['Close']
close_prices1.head()
```

```python
# Close prices of all the macro assets
close_prices = full_history_data['Close']

display(close_prices.head())
display(close_prices.tail())

# Save close prices to CSV
close_prices.to_csv('../macro_data/close_prices.csv')
```

```python
%pip install pandas
%pip install numpy

import pandas as pd
import numpy as np
from scipy.stats import gmean

macro_assets = close_prices.copy()

# Reset index to convert 'Date' from index to column
macro_assets.reset_index(inplace=True)

# Convert 'Date' column to datetime
macro_assets['Date'] = pd.to_datetime(macro_assets['Date'], format='%Y-%m-%d')

# Set 'Date' column back as index
macro_assets.set_index('Date', inplace=True)

macro_assets
```

```python
macro_assets['SPY'].plot(figsize=(12, 6), legend=True)
macro_assets['TLT'].dropna().plot(figsize=(12, 6), legend=True)
macro_assets['GLD'].dropna().plot(title='Macro Assets Close Prices', figsize=(12, 6), legend=True)
```

```python
macro_assets.replace(np.nan, 0, inplace=True)
macro_assets
```

```python
macro_asset_returns = macro_assets.pct_change()
macro_asset_returns = macro_asset_returns.drop(macro_asset_returns.index[0])

macro_asset_returns
```

```python
macro_asset_monthly_returns = macro_asset_returns.resample('M').apply(lambda x: (x + 1).prod() - 1).reset_index()

macro_asset_monthly_returns['Date'] = macro_asset_monthly_returns['Date'].dt.strftime('%Y-%m')

macro_asset_monthly_returns.set_index('Date', inplace=True)

display(macro_asset_monthly_returns)

spy = macro_asset_monthly_returns['SPY']
tlt = macro_asset_monthly_returns['TLT']
gld = macro_asset_monthly_returns['GLD']

spy.replace([np.inf, -np.inf], 0, inplace=True)
tlt.replace([np.inf, -np.inf], 0, inplace=True)
gld.replace([np.inf, -np.inf], 0, inplace=True)
```

```python
spy_start_date = spy[spy != 0].index.min()
tlt_start_date = tlt[tlt != 0].index.min()
gld_start_date = gld[gld != 0].index.min()
```

```python
filtered_spy = spy[spy_start_date:]
filtered_tlt = tlt[tlt_start_date:]
filtered_gld = gld[gld_start_date:]
```
